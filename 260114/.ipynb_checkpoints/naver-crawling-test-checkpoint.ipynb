```python
# 네이버 웹포털 사이트 내 뉴스카테고리에서 특정키워드 크롤링 해오기
# 크롤링 (crawling) : 외부에서 데이터를 수집해오는 행위를 의미함
# ctrl + / : 주석처리 가능, ctrl + z : 직전단계로 돌아가기

import requests
from bs4 import beautifulsoup

query = "마케팅"
url = f"https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={query}" # f-string 문법

headers = {
    "user-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = beautifulsoup(response.text, "html.parser")

titles = soup.select("a.fender-ui_228e3bd1,MNH2zYqVX0VCatL92VdS")

for i, title in enumerate(titles[0:10], 1) :
    print(f"{i}, {title.text}")

# bs4.BeautifulSoup()
# print(response) #print 함수 => 어떤 기능을 구현하고자 할 때
# 방법은 여러가지!! 

# 기초의 소중함
# 학습 : 기초 지식 + 생성형 AI => 더 많은 구현
# 응용 & 문제 해결 능력
# 변수, 문자열 자료형, 디셔너리 자료구조, 리스트 자료구조, 반복문, enum타입, 인덱스, 슬라이싱 기능 출력?
```
