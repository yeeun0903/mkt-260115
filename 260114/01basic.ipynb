# 파이썬 기초문법
- 출력 : print() > 내장함수 = built in function / 내장 : 애초에 파이선 문법 탄생될 때부터 존재
- 입력 : input("당신의 이름은?") > 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
   - 해당함수를 통해서 수집한 값은 무조건 "문자열"이 된다는 사실을 기억해라!
- 연산 :
   - 사칙연산 : + @ → + - * / %(나누기 연산에 대한 나머지) ** (제곱) ※ '+' : 문자간의 연결연산자로도 사용가능!
   - 비교연산 : <, >, <=, >=, ==(같다), !=(같지 않다) ※ (좌항 = 우항, 우항의 값을 좌항에 삽입&적용)
- 변수 :
   - variable : 변할 수 있는 값은 넣는 공간 = 그릇
   - 변수명 작업 : 직관, 명시 → 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 먼저올 수 없음 (중간에 오는 것은 가능)
   - 팁 : 스네이크 표기법 (뱀 → case_study / 카멜 표기법 → caseStudy)
- 자료형 : 변수 안에 담을 수 있는 값의 형태 → 문자열(String)"" VS 문자 (Character)'' / 정수(Int) / 실수(Float) / 논리형(Boolean)
   - 형변화 : 자동형변환 → input() / 명시적형변환 → 수동적으로 형태를 변환
   - 자료형확인 : type()
   - int() : 정수형 변환
   - float() : 실수형 변환
   - str() : 문자열 변환
   - bool() : 논리형 변환
   - list() : 리스트 자료구조 변환
   - dict() : 딕셔너리 자료구조 변환
   - tuple() : 튜플 자료구조 변환
   - set() : 세트(집합) 자료구조 변환
- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
   - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대, 소문자 구분)

```python
print("Hello World")
print('H')
print(1+2)
print((1+2)*3)
```

```python
num1 = 4 # 변수 : 어떤 값을 담을 수 있는 그릇 → 값을 언제든지 바꿔서 담을 수 있음.
num2 = 2

# num1 = 8 # 변수 안의 값을 "재할당"
print(num1+num2) # 4 , 2 → 피연산자 / + → 연산자
print(num1-num2)
print(num1*num2)
print(num1/num2) # 나누기 연산에 대한 몫
print(num1%num2) # 나누기 연산에 대한 나머지 → 홀수, 짝수를 찾아내야 하는 상황
print(num1**num2)
```

```python
str01 = "David"
str02 = 'Jane'
int01 = 1
int02 = "1"
flo01 = 3.14
bool01 = True
bool02 = False

print(str01, str02, int01, int02)
print(type(int02)) # type 함수 : 인자값으로 입력된 변수 안에 담긴 자료의 타입, 형태를 식별&확인하고자 할 때 함수
print(type(flo01)) # alt + shift + 아래방향 화살표 = duplicate 복제됨!
print(type(int01)) 
print(type(bool01)) 
```

```python
# 주피터 노트북 초심자 좋은 이유 : 실제 코드 작성 외 굳이 외적인 불필요한 작업 X
# 주피터 사용하지 않을 때 → scrapy 기법 : cursor AI 
print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
print(num1 == num2)
print(num1 != num2)
```

```python
radius = 10
pi = 3.1415

# 변수를 선언해서 다음의 값을 출력해주세요
# 지름의 값
# 원의 둘레 : 2 파이 r
# 원의 넓이 : 파이 r 제곱
# 각자 변수 명을 선언 → 연산식을 활용해서 위의 값을 도출!

radius2 = radius*2
radius3 = 2*pi*radius
radius4 = pi*radius**2
print(radius2,radius3,radius4)
print("지름 : ", radius2)
print("원의 둘레 :", radius3)
print("원의 넓이 :", radius4)

print(f"원의 넓이 : {radius4}")
```
