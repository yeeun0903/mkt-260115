# git 사용법
1) 로컬 컴퓨터 (내 컴퓨터) : git 응용 프로그램 다운 및 설치
2) 터미널 : git -v 명령어
   - 검색바 > cmd > 창 클릭 : 검정색 창
3) 만약, 없다면 > 구글 크롬 > git 다운로드 후 설치
   - homebrew : 대리로 응용프로그램 다운로드 후 설치해주는 중간 프로그램 (맥용)
   - chocolatey : 대신 설치해주는 곳 (윈도우용)
4) git의 소유주 확인 & 변경
   - git config --list
5) github 회원가입 : 계정정보 값으로 입력
   - git : 나의 로컬컴퓨터에서 버전 관리 툴
   - github : 나의 로컬컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할 때, 혹은 내가 사용하고 있는 디바이스가 2대 이상일 때 (A,B 기기 모두에서 데이터를 관리하고 싶을 때)
   - repository : github > project에 따른 개별 폴더
   - user.email : yeaeun0@gmail.com 
   - user.name : yeeun0903
   - 터미널(윈도우) 재접속 시 > 관리자모드 : 전역(글로벌)세팅
   - git config --global user.name "yeeun0903"
   - git config --global user.email "yeaeun0@gmail.com"
6) git commit & push
   - cd + "현재 나의 루트폴더" 이동 (파일 위치 주소 복붙)
   - git 초기화하기 : git init
   - staging area : 데이터 전송  git add .
   - 데이터 tracking : git status
   - 데이터 labeling : git commit -m "260115"
   - 원격(웹) github : git remote add origin https://github.com/yeeun0903/mkt-260115.git (< 레포지토리 주소임!)
   - 로컬(git) & 원격(github) 연결 잘 됐는지 확인법 : git remote -v
   - 로컬 > 원격으로 데이터 push : git push origin master
7) git clone & pull
   - 개인 노트북에 git 설치완료!
   - 집 도착 > 노트북 on : 새폴더 생성(Root 폴더) > 폴더 주소 복사 > 터미널 들어가기 > cd + 주소 붙여넣기
   - 해당 Root 폴더 : git clone
   - 집에서 학원파일 오픈 후 추가로 업데이트 했을 때 > 다시 내 레포지토리 push
   - git add . < 여기부터 시작
8) git fork
   - https://github.com/Divjason/mkt-260115.git
