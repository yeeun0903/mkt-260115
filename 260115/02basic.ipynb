# 파이썬 기초문법
- 출력 : print() > 내장함수 = built in function / 내장 : 애초에 파이선 문법 탄생될 때부터 존재
- 입력 : input("당신의 이름은?") > 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
   - 해당함수를 통해서 수집한 값은 무조건 "문자열"이 된다는 사실을 기억해라!
- 연산 :
   - 사칙연산 : + @ → + - * / %(나누기 연산에 대한 나머지) ** (제곱) ※ '+' : 문자간의 연결연산자로도 사용가능!
   - 비교연산 : <, >, <=, >=, ==(같다), !=(같지 않다) ※ (좌항 = 우항, 우항의 값을 좌항에 삽입&적용)
- 변수 :
   - variable : 변할 수 있는 값은 넣는 공간 = 그릇
   - 변수명 작업 : 직관, 명시 → 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 먼저올 수 없음 (중간에 오는 것은 가능)
   - 팁 : 스네이크 표기법 (뱀 → case_study / 카멜 표기법 → caseStudy)
- 자료형 : 변수 안에 담을 수 있는 값의 형태 → 문자열(String)"" VS 문자 (Character)'' / 정수(Int) / 실수(Float) / 논리형(Boolean)
   - 형변화 : 자동형변환 → input() / 명시적형변환 → 수동적으로 형태를 변환
   - 자료형확인 : type()
   - int() : 정수형 변환
   - float() : 실수형 변환
   - str() : 문자열 변환
   - bool() : 논리형 변환
   - list() : 리스트 자료구조 변환
   - dict() : 딕셔너리 자료구조 변환
   - tuple() : 튜플 자료구조 변환
   - set() : 세트(집합) 자료구조 변환
- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
   - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대, 소문자 구분)
   - len() : length = 길이
   - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇 번째 인덱스에 위치하는지 찾고자 할 때
   - replace() : 특정 문자열을 대체하고자 할 때
   - index : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있다 > 문자열 왼쪽부터 숫자 시작 : 0부터 시작
   - slicing : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때
   - strip() : 특정 문자열 앞, 뒤쪽에 여백 혹은 여백으로 인지할 수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할 때
   - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때, 사용할 수 있는 기능 (메서드 함수)
- 형식지정자 :
   - 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 찾아와서 출력 결정
   - %s - string
   - %c - character
   - %d - int
   - %f - float

```python
article = """Who would have thought serums infused with snail mucin - the sticky substance they secrete - would become a part of skincare routines around the world?

Well, it's happened - and the gooey elasticity is key, according to a viral TikTok challenge promoting the serum. It made its manufacturer, the small South Korean label CosRX, go global. It is now owned by Amorepacific, the country's biggest cosmetics company.

The rapid spread of that sticky serum tells you just how wildly successful K-beauty has become. Fuelled by viral content and trends, it is one of the biggest industries in South Korea, where the pressure to look almost flawless has always been huge in a highly competitive society."""

print(article.count("serum"))
print(len(article))
print(article.find("o")) # 문서를 읽어가면서 가장 처음 찾은 문자의 인덱스 변환
print(article.replace("Who", "When"))
```

```python
text = "banana"

index = text.find("a") # 1,3,5

next_index = text.find("a", 3 + 1)

print(index, next_index)
```

```python
# 반복문 사용

text = "banana"
target = "a"

idx = text.find(target) # 존재하지 않는 문자를 찾아오라고 할 때 : -1

# 예외 조항 처리 : 어떤 명령문을 실행시키고자 할 때, 만약 ~가 아니라면, 해라!

while idx != -1 :
    print(idx)
    idx = text.find(target, idx + 1)
```

```python
text = "python"

print(text)
print(text[0])
print(text[1])
# 어떤 문자열이던, 마지막번째 문자의 인덱스 넘버는 전체 문자열 개수 -1 과 동일

print(len(text))
print(text[len(text) - 1])
print(text[-2])

# 특정 사이트 접속 > 리뷰 크롤링 > 댓글 
```

```python
text = "python"

print(text)
print(text[2])
print(text[2:4]) # 찾아오고자하는 문자열[start:finish + 1]
print(text[2:6]) 
print(text[2:]) # slicing에서 마지막 번째 인덱스의 생략 = 끝까지를 의미함!
print(text[2:len(text)]) # 해당 구문을 가장 많이 사용함
print(text[2:-1])
```

```python
name = "I have a {}, I have an {}."
print(name)

print(name.format("pen","apple"))
print("I have a {}, I have an {}.".format("pen","apple"))
print("I have an {1}, I have a {0}.".format("pen","apple")) # 객체 - 메서드(method) 함수 -> 앞에 온점이 있으면 메서드 함수
# 문자열 형식 객체.format()
```

```python
interest = 0.087
print(format(interest, ".2f")) # 독립함수(문자 & 숫자 X) -> 소수점 위치 일치
```

```python
# 파이썬 내장함수 가운데, 소수점 자리를 올리거나, 내리는 독립 기능은 없다!
# 외부 라이브러리 모듈 찾아와서 사용!!
# 수학객체 라이브러리 모듈

import math

interest = 0.891
result01 = math.floor(interest) # 바닥 -> 버리다
print(result01) 

result02 = math.ceil(interest) # 천장 -> 올리다
print(result02) 
```

```python
print("I have a %f, I have an %d." % (3.14, 7))
print("I have a %f, I have an %c." % (3.14, 'A')) # 엄격한 프로그래밍 언어 -> "", ''
print("I have a %f, I have an %c." % (3.14, 65)) # 문자와 정수는 치환가능하다! ASCII Chart 참고
print("I have a %f, I have an %s." % (3.14, "AB")) 
```
