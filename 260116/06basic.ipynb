# 파이썬 기초문법
- 출력 : print() > 내장함수 = built in function / 내장 : 애초에 파이선 문법 탄생될 때부터 존재
- 입력 : input("당신의 이름은?") > 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
   - 해당함수를 통해서 수집한 값은 무조건 "문자열"이 된다는 사실을 기억해라!
- 연산 :
   - 사칙연산 : + @ → + - * / %(나누기 연산에 대한 나머지) ** (제곱) ※ '+' : 문자간의 연결연산자로도 사용가능!
   - 비교연산 : <, >, <=, >=, ==(같다), !=(같지 않다) ※ (좌항 = 우항, 우항의 값을 좌항에 삽입&적용)
   - (복합)대입연산 :  
- 변수 :
   - variable : 변할 수 있는 값은 넣는 공간 = 그릇
   - 변수명 작업 : 직관, 명시 → 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 먼저올 수 없음 (중간에 오는 것은 가능)
   - 팁 : 스네이크 표기법 (뱀 → case_study / 카멜 표기법 → caseStudy)
- 자료형 : 변수 안에 담을 수 있는 값의 형태 → 문자열(String)"" VS 문자 (Character)'' / 정수(Int) / 실수(Float) / 논리형(Boolean)
   - 형변화 : 자동형변환 → input() / 명시적형변환 → 수동적으로 형태를 변환
   - 자료형확인 : type()
   - int() : 정수형 변환
   - float() : 실수형 변환
   - str() : 문자열 변환
   - bool() : 논리형 변환
   - list() : 리스트 자료구조 변환
   - dict() : 딕셔너리 자료구조 변환
   - tuple() : 튜플 자료구조 변환
   - set() : 세트(집합) 자료구조 변환
- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
   - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대, 소문자 구분)
   - len() : length = 길이
   - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇 번째 인덱스에 위치하는지 찾고자 할 때
   - replace() : 특정 문자열을 대체하고자 할 때
   - index : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있다 > 문자열 왼쪽부터 숫자 시작 : 0부터 시작
   - slicing : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때
   - strip() : 특정 문자열 앞, 뒤쪽에 여백 혹은 여백으로 인지할 수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할 때
   - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때, 사용할 수 있는 기능 (메서드 함수)
- 형식지정자 :
   - 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 찾아와서 출력 결정
   - %s - string
   - %c - character
   - %d - int
   - %f - float

- a = a + b -> a += b
- a = a - b -> a -= b
- a = a * b -> a *= b
- a = a / b -> a /= b
- a = a % b -> a %= b
- a = a ** b -> a **= b

```python
sum = 0 # 아무런 값에 영향을 미치지 않는 값 = 0

for index in range(1, 11) : # 첫번째 인자값 ~ 두번째 인자값 -1까지 : 1 ~ 10
    # sum = sum + index #누산기
    sum += index # 복합대입 연산자

print(sum)
```

```python
name = ""

while name != "David" : 
    name = input("당신의 이름은?")
    if name != "David" :
        print (f"Hi~{name}, You know David?")
    else :
        print(f"Hi~ {name}")
```

```python
# 사용자에게 1개의 숫자를 받아주세요.

num = int(input("몇 단의 구구단을 출력할까요?"))

for digit in range(1, 10) : 
    print(f"{num} x {digit} = {num * digit}")
```

```python
# 1 ~ 30까지의 숫자 (정수) 중 3의 배수만 출력해주세요.

num = 0

while num < 30 : 
    num += 1
    if num % 3 == 0 :
        print(num)
```
