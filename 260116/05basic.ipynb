```python
# 사용자로부터 값을 받아서 짝수인지, 홀수인지 판단 후 짝수라면, "짝수입니다." 출력 / 홀수라면, "홀수입니다."

number = int(input("현재 가지고 있는 숫자는?"))

if number % 2 == 0 :
    print("짝수입니다.")
else :
    print("홀수입니다.")
```

```python
# 사용자에게 3개의 숫자를 입력받은 후 그 중 작은 숫자를 출력하세요.

num01 = int(input("첫번째 숫자"))
num02 = int(input("두번째 숫자"))
num03 = int(input("세번째 숫자"))

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
elif num03 <= num01 and num03 <= num02 :
    print(num03)
```

```python
numbers = []

for i in range(3) : 
    number = int(input(f"{i+1}번째 숫자 : "))
    numbers.append(number)

# 구조분해할당 문법 : 리스트와 같이 여러개의 변수값을 가지고 있는 자료구조의 값들을 하나씩 추출 > 독립변수로 담을 수 있는 문법
num01, num02, num03 = numbers

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
elif num03 <= num01 and num03 <= num02 :
    print(num03)
```

```python
numbers = []

for i in range(3) : # 반복(문 : 식 + 식)
    number = int(input(f"{i+1}번째 숫자 : "))
    numbers.append(number)

print(min(numbers))
```

```python
# 리스트 컴프리헨션 = list comprehension
nums = [int(input(f"{i + 1}번째 숫자 : ")) for i in range(3)]
print(min(nums))
```

```python
# 리스트 컴프리헨션 = list comprehension

nums = [int(input(f"{i + 1}번째 숫자 : ")) for i in range(3)]
# 리스트의 형태를 가지고 있지만, 실제 리스트의 속성 값을 모두 사용할 수는 없음.
# sort, reverse -> 원래 값을 편집 -> 인덱스 구조 x

print(nums)
print(sorted(nums)[0]) # 완전 새로운 리스트 구조 -> 인덱스 o

# print(min(nums))
```
